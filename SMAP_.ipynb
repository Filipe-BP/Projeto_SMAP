```python
import pandas as pd
dados = pd.read_csv('smap_dados.csv')
qcai_lista = []
ndias = 685
ad = 1300
tuin = 42.06
ebin = 0.228
str = 355.9164
k2t = 1.17226577269563
crec = 0.972941542777403
ai = 4
capcc = 50
kkt = 30
pcof = 1


capc =capcc/100
crec = crec/100
kk = 0.5 ** (1/kkt)
k2t = 0.5 ** (1/k2t)
rsolo = tuin / 100 * str
rss = ebin / (1 - kk) / ad * 86.4
res = 0
tuin = tuin/100



for i in range(ndias):

    ep = dados['evaporacao'][i]
    pri = dados['precipitacao'][i]

    pre = pri * pcof
    tu = rsolo / str
    if(pre > ai):
        es = (pre - ai)**2 / (pre - ai + str - rsolo)
    else:
        es = 0


    if((pre-es) > ep):
        er = ep
    else:
        er = (pre - es) + (ep -(pre - es)) * tu


    if(rsolo > (capc * str)):
        rec = crec * tu * (rsolo-(capc * str))
    else:
        rec = 0


    rsolo = rsolo + pre - es - er - rec
    if(rsolo > str):
        es = es + rsolo - str
        rsolo = str


    ed = (res + es) * (1 - k2t)
    res = res + es - ed
    eb = rss * (1 - kk)
    rss = rss - eb + rec
    qcai = (ed + eb) * ad / 86.4
    print(f' capc={capc}, crec={crec}, kk={kk}, k2t={k2t}, rsolo={rsolo}, res={res}, ed={ed}, es={es}, eb={eb}, rss={rss}, qcai={qcai}')
    qcai_lista.append(qcai)

print(qcai_lista)
qcai_lista = pd.DataFrame(qcai_lista, columns = ['qcai'])
qcai_lista.to_csv('qcai_1.csv', index = False)
```
